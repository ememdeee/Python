```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from pytesseract import image_to_string

img1 = mpimg.imread('tulisan.png')
print("Original image")
plt.imshow(img1, cmap="gray")
plt.show()

img = cv2.imread('tulisan.png')
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print("Grayscale image")
plt.imshow(img_gray, cmap='gray')
plt.show()



text = image_to_string(img_gray)  
print(text)



```

```
img3 = cv2.imread('imggray.png')
ret, img_binary = cv2.threshold(img3, 120, 255, cv2.THRESH_BINARY)
plt.imshow(img_binary)
plt.show()

text = image_to_string(img_binary)
print(text)
```
