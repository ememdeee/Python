```python
import requests
from bs4 import BeautifulSoup

judul_array=[]
kurensy_array=[]
harga_array=[]
katgori_array=[]
stok_array=[]

for i in range (1):
  url="https://books.toscrape.com/catalogue/page-"+str(i+1)+".html"   #next url:https://books.toscrape.com/catalogue/page-2.html
  r=requests.get(url)
  soup = BeautifulSoup(r.content,'html.parser')
  items=soup.find_all("h3")
  # print(soup)

  for item in items:
    url_main= "https://books.toscrape.com/catalogue/"+item.find ('a')['href']
    print (url_main)
    r2=requests.get(url_main)
    soup2 = BeautifulSoup(r2.content,'html.parser')
    judul=soup2.find("h1") #judul
    kategories=soup2.find_all("li",limit=3) #kategori
    # kategories
    for kategori in kategories: #ngeprint 3 "li", lalu di for agar yg diambil yang terakhir
      s=1+1
    kategori=kategori.text[1:-1]
    harga=soup2.find("p",{"class":"price_color"}) #harga
    kurensy=harga.text[0] #ambil character awal
    harga_saja=harga.text[1:] #hapus character awal

    stocks=soup2.find("p",{"class":"instock"}) #stock
    a=stocks.text
    arr = list(a)
    stocky=arr[25]+arr[26]

    #input array
    judul_array.append(judul.text)
    kurensy_array.append(kurensy)
    harga_array.append(harga_saja)
    katgori_array.append(kategori)
    stok_array.append(stocky)
    
    print(judul.text,kurensy,harga_saja,kategori,stocky)
```

```python
import csv
f=open ('list buku.csv','w')
writer = csv.writer(f)

writer.writerow(['judul','currency','harga','kategori','stock'])
for i in range (0, len(judul_array)):
  writer.writerow([judul_array[i],kurensy_array[i],harga_array[i],katgori_array[i],stok_array[i]])
f.close()
```
