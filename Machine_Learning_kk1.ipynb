```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import csv

#Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Data Mining/sberbank.csv')

dataset_cat= pd.DataFrame()
for i in list(dataset.columns):
  dataset_cat['{}_cat'.format(i)] = dataset[i].astype('category').copy()
  dataset_cat['{}_cat'.format(i)] = dataset_cat['{}_cat'.format(i)].cat.codes

print (dataset_cat.head())

print (dataset_cat.isnull().sum())

print(dataset_cat.describe().loc[['count','mean','std','min','max']])

Y = np.array(dataset_cat['Kategori_cat'])
X = dataset_cat.drop('Kategori_cat',axis=1)

#print(Y)
 
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.30)

model = RandomForestClassifier(n_estimators= 10000, random_state = 24)

model.fit(X_train, y_train)

y_pred= model.predict(X_train)
y_pred1 = model.predict(X_test)

print('in-sample accuracy')
from sklearn.metrics import classification_report
print(classification_report(y_train, y_pred))

print('out-sample accuracy')
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
import csv

#Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Data Mining/sberbank.csv')

dataset_cat= pd.DataFrame()
for i in list(dataset.columns):
  dataset_cat['{}_cat'.format(i)] = dataset[i].astype('category').copy()
  dataset_cat['{}_cat'.format(i)] = dataset_cat['{}_cat'.format(i)].cat.codes

print (dataset_cat.head())

print (dataset_cat.isnull().sum())

print(dataset_cat.describe().loc[['count','mean','std','min','max']])

Y = np.array(dataset_cat['Kategori_cat'])
X = dataset_cat.drop('Kategori_cat',axis=1)

#print(Y)
 
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.30)

model = CategoricalNB()

model.fit(X_train, y_train)

y_pred= model.predict(X_train)
y_pred1 = model.predict(X_test)

print('in-sample accuracy')
from sklearn.metrics import classification_report
print(classification_report(y_train, y_pred))

print('out-sample accuracy')
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```
