```python
import requests
from bs4 import BeautifulSoup
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

#melihat ke unikan dan menampilkan plot jika perlu
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

```python
url="https://best-hashtags.com/best-hashtags.php"
r = requests.get(url) #akan request dan mengambil data yang ada di web. semua nya beserta iklan2
print(r.status_code) #ngeprint code masuknya bisa atau g, kalau hasil nya 200 berhasil di request dan data uda mask di r semua, kalau 404 not faund

soup = BeautifulSoup(r.content,'html.parser')
for i in range(6, 106):  #jumlah i nya itu ada 100, dan kenapa dimuali dari 6? itu adalah trik biasa untuk melompati a ke 6 karena disitu yang ada a href 
  banyaklink=soup.find_all("a",limit=i) #mendapatkan link a href:""
  banyaklink
  for link in banyaklink:
    b=link

  url_main= "https://best-hashtags.com"+b.get('href')
  print(url_main)
  r = requests.get(url_main)
  if (r.status_code==200): #melihat link nayala atau tidak
    soup2 = BeautifulSoup(r.content,'html.parser')
    hastag1=soup2.find("p1")
    hastag2=soup2.find("p2")

    a=hastag1
    b=hastag2
    #merubah tipe nya hatag1,2 yang seblumnya bs4.element.Tag karena hasil dari scrap, skrg jadi list. Kalau mau ke string tinggal panggil paragraphs[0] dan seterusnya
    paragraphs = [] #bakal ada  data 200
    for hastag in hastag1:
      paragraphs.append(str(hastag))
    for hastag in b:
      paragraphs.append(str(hastag))

    #memisahkan setiap katanya dengan tekonized yang berasal dari array paragraph 0 dan 1  
    for i in range(2):
      tokenized_word=word_tokenize(paragraphs[i]) #dijalankan 2x, buat [0] dan [1] dari array paragraphs.->
      semua_hastag=semua_hastag+tokenized_word
  else:
    print("link diatas mati")
```

```python
#menggunakan stop word untuk menghapus # dari dari semua_hastag
stop_words=set({'#'}) #ini bisa diatur atur kalo perlu bhfyp juga dihapus
print("yang dihapus:",stop_words)
filtered_hastag=[] #clearing data dari semua_hastag
for w in semua_hastag:
    if w not in stop_words:
        filtered_hastag.append(w)
print("Tokenized Sentence:",semua_hastag)
print("panjang array:",len(semua_hastag))
print("Filterd Sentence:",filtered_hastag)
print("panjang array:",len(filtered_hastag))
```

```python
#melihat kesamaan atau grafiknya
#mungkin tidak perlu karena akan dilakukan di laravel yang kurang hanya pengolahan datanya

fdist = FreqDist(filtered_hastag) #fdist itu banyak hastag yang berbeda, jadi 1112 itu jumlah hastag yang unik sisanya hastag yang sama.
print(fdist)

fdist.plot(40,cumulative=False)
plt.show()
```
